**Задание 4.2.2**

Для функции sin(𝑥𝑒^cos(𝑥)) на интервале [0;10] с шагом 0,05 построить график функции, закрасить площадь под ним и вычислить значение этой площади.

```python
#Задание 4.2.2

import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz
x = np.arange(0.0, 10, 0.1)
y = np.abs(np.sin(x*np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c = 'r')
plt.fill_between(x, y)
area = trapz(y)
print(area)
```

**Задание 4.3.2**

Дана функция (не получится вставить). Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива – линейный и для среза – точечный 

```python
#Задание 4.3.2

import matplotlib.pyplot as plt
from math import *

x1 = range(1, 11)
x2 = []
x3 = range(1, 6)
x4 = []

for x in x1:
    y = (((1 + exp(x ** 0.5) + cos(x) ** 2) ** 0.5) / fabs(1 - sin(x) ** 3)) + log(2 * x, exp(1))
    x2.append(y)
for x in range(len(x2)//2):
    x4.append(x2[x])

plt.figure(figsize=(9, 9))

# График для основного массива
plt.subplot(2, 1, 1)
plt.plot(x1, x2, label="Основной массив")
plt.ylabel("Основной массив", fontsize=14)
plt.grid()
plt.legend()

# График для среза
plt.subplot(2, 1, 2)
plt.scatter(x3, x4, label="Срез", marker='o')
plt.ylabel("Срез", fontsize=14)
plt.grid()
plt.legend()

plt.show()
```

**Задание 4.3.3**

Дана функция |cos(𝑥𝑒^(cos(𝑥)+ln(𝑥+1)))| Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним.

```python
#Задание 4.3.3

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(0, 10 ,1)
y = np.fabs(np.cos(x*(np.exp(np.cos(x) + np.log(x+1)))))

plt.grid()
plt.plot(x, y, c = 'black')
plt.fill_between(x, y)
area = trapz(y)
print(area)
```

**Задание 4.3.4**

Откройте статистические данные по стоимости акций Apple, Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на плоскости и оцените их динамику. Подумайте, как можно улучшить читаемость результатов. 

```python
#Задание 4.3.4

import matplotlib.pyplot as plt

# Данные по стоимости акций за 12 месяцев 2021 года
stocks = {
    "Apple": [131.96, 124.61, 136.96, 124.47, 152.83, 167.29, 170.34, 165.12, 174.01, 156.71, 148.8, 136.72, 149.9],
    "Microsoft": [222.53, 235.06, 235.90, 238.47, 254.40, 251.21, 269.51, 286.36, 302.87, 282.12, 331.36, 335.15, 310.41],
    "Google": [87.8, 96.2, 102.3, 104.90, 120.14, 121.10, 124.85, 135.48, 145.65, 133.55, 148.16, 144.21, 144.48]
}

total_prices = {stock: sum(prices) for stock, prices in stocks.items()}

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(total_prices.values(), labels=total_prices.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Суммарная стоимость акций за 12 месяцев 2021 года')
plt.show()

# Построение графика стоимости акций
plt.figure(figsize=(10, 6))
for stock, prices in stocks.items():
    plt.plot(prices, label=stock)

plt.xlabel("Месяцы года")
plt.ylabel("Стоимость акций")
plt.title('Динамика стоимости акций за 12 месяцев 2021 года')
plt.legend()
plt.grid()
plt.show()
```

**Задание 4.3.5**

Создайте простейший калькулятор, включающий основные действия для двух переменных ′+′, ′−′, ′∙′, ′÷′, а также вычисление следующих функций: 𝑒^𝑥+𝑦, sin(𝑥+𝑦), cos(𝑥+𝑦), 𝑥^𝑦

```python
#Задание 4.3.5

from tkinter import *
from math import exp, sin, cos

def calculate(operator):
    x = txt1.get()
    y = txt2.get()

    try:
        x = float(x)
        y = float(y)

        if operator == '+':
            res = x + y
        elif operator == '-':
            res = x - y
        elif operator == '*':
            res = x * y
        elif operator == '/':
            res = x / y
        elif operator == 'exp':
            res = exp(x) + y
        elif operator == 'sin':
            res = sin(x + y)
        elif operator == 'cos':
            res = cos(x + y)
        elif operator == '^':
            res = x ** y
        else:
            res = "Неверный оператор"

        result = "Результат - {}".format(res)
        lbl.configure(text=result)
    except ValueError:
        lbl.configure(text="Ошибка: Введите корректные числа")

window = Tk()
window.title("Калькулятор")
window.geometry('400x300')

# Определение стилей
entry_style = {'width': 10, 'font': ('Arial', 12)}

# Ввод значений x и y
lblx = Label(window, text="x", font=('Arial', 14))
lblx.grid(column=0, row=0, pady=10)
txt1 = Entry(window, **entry_style)
txt1.grid(column=1, row=0, pady=10)
lbly = Label(window, text="y", font=('Arial', 14))
lbly.grid(column=2, row=0, pady=10)
txt2 = Entry(window, **entry_style)
txt2.grid(column=3, row=0, pady=10)

# Результат
lbl = Label(window, text="Результат - _", font=('Arial', 14))
lbl.grid(column=0, row=1, columnspan=4, pady=10)

# Кнопки с операторами
operators = ['+', '-', '*', '/', 'exp', 'sin', 'cos', '^']
row_number = 2
for operator in operators:
    if operator == 'sin' or operator == 'cos':
        btn_text = f"{operator}(x+y)"
    else:
        btn_text = f"x {operator} y"
    btn = Button(window, text=btn_text, command=lambda op=operator: calculate(op), font=('Arial', 12))
    btn.grid(column=operators.index(operator) % 4, row=row_number, padx=5, pady=5)
    if operators.index(operator) % 4 == 3:
        row_number += 1

# Запуск приложения
window.mainloop()
```
