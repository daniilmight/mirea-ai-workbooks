**–ó–∞–¥–∞–Ω–∏–µ 4.3.2**

–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è (–Ω–µ –ø–æ–ª—É—á–∏—Ç—Å—è –≤—Å—Ç–∞–≤–∏—Ç—å). –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π 

```python
#–ó–∞–¥–∞–Ω–∏–µ 4.3.2

import matplotlib.pyplot as plt
from math import *

x1 = range(1, 11)
x2 = []
x3 = range(1, 6)
x4 = []

for x in x1:
    y = (((1 + exp(x ** 0.5) + cos(x) ** 2) ** 0.5) / fabs(1 - sin(x) ** 3)) + log(2 * x, exp(1))
    x2.append(y)
for x in range(len(x2)//2):
    x4.append(x2[x])

plt.figure(figsize=(9, 9))

# –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞
plt.subplot(2, 1, 1)
plt.plot(x1, x2, label="–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤")
plt.ylabel("–û—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤", fontsize=14)
plt.grid()
plt.legend()

# –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è —Å—Ä–µ–∑–∞
plt.subplot(2, 1, 2)
plt.scatter(x3, x4, label="–°—Ä–µ–∑", marker='o')
plt.ylabel("–°—Ä–µ–∑", fontsize=14)
plt.grid()
plt.legend()

plt.show()
```

```python
#–ó–∞–¥–∞–Ω–∏–µ 4.3.3

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(0, 10 ,1)
y = np.fabs(np.cos(x*(np.exp(np.cos(x) + np.log(x+1)))))

plt.grid()
plt.plot(x, y, c = 'black')
plt.fill_between(x, y)
area = trapz(y)
print(area)
```

```python
#–ó–∞–¥–∞–Ω–∏–µ 4.3.4

import matplotlib.pyplot as plt

# –î–∞–Ω–Ω—ã–µ –ø–æ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∞–∫—Ü–∏–π –∑–∞ 12 –º–µ—Å—è—Ü–µ–≤ 2021 –≥–æ–¥–∞
stocks = {
    "Apple": [131.96, 124.61, 136.96, 124.47, 152.83, 167.29, 170.34, 165.12, 174.01, 156.71, 148.8, 136.72, 149.9],
    "Microsoft": [222.53, 235.06, 235.90, 238.47, 254.40, 251.21, 269.51, 286.36, 302.87, 282.12, 331.36, 335.15, 310.41],
    "Google": [87.8, 96.2, 102.3, 104.90, 120.14, 121.10, 124.85, 135.48, 145.65, 133.55, 148.16, 144.21, 144.48]
}

total_prices = {stock: sum(prices) for stock, prices in stocks.items()}

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –∫—Ä—É–≥–æ–≤–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã
plt.figure(figsize=(8, 8))
plt.pie(total_prices.values(), labels=total_prices.keys(), autopct='%1.1f%%', startangle=90)
plt.title('–°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∞–∫—Ü–∏–π –∑–∞ 12 –º–µ—Å—è—Ü–µ–≤ 2021 –≥–æ–¥–∞')
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∞–∫—Ü–∏–π
plt.figure(figsize=(10, 6))
for stock, prices in stocks.items():
    plt.plot(prices, label=stock)

plt.xlabel("–ú–µ—Å—è—Ü—ã –≥–æ–¥–∞")
plt.ylabel("–°—Ç–æ–∏–º–æ—Å—Ç—å –∞–∫—Ü–∏–π")
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∞–∫—Ü–∏–π –∑–∞ 12 –º–µ—Å—è—Ü–µ–≤ 2021 –≥–æ–¥–∞')
plt.legend()
plt.grid()
plt.show()
```

**–ó–∞–¥–∞–Ω–∏–µ 4.3.5**

–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è –¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤+‚Ä≤, ‚Ä≤‚àí‚Ä≤, ‚Ä≤‚àô‚Ä≤, ‚Ä≤√∑‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^ùë•+ùë¶, sin(ùë•+ùë¶), cos(ùë•+ùë¶), ùë•^ùë¶

```python
#–ó–∞–¥–∞–Ω–∏–µ 4.3.5

from tkinter import *
from math import exp, sin, cos

def calculate(operator):
    x = txt1.get()
    y = txt2.get()

    try:
        x = float(x)
        y = float(y)

        if operator == '+':
            res = x + y
        elif operator == '-':
            res = x - y
        elif operator == '*':
            res = x * y
        elif operator == '/':
            res = x / y
        elif operator == 'exp':
            res = exp(x) + y
        elif operator == 'sin':
            res = sin(x + y)
        elif operator == 'cos':
            res = cos(x + y)
        elif operator == '^':
            res = x ** y
        else:
            res = "–ù–µ–≤–µ—Ä–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä"

        result = "–†–µ–∑—É–ª—å—Ç–∞—Ç - {}".format(res)
        lbl.configure(text=result)
    except ValueError:
        lbl.configure(text="–û—à–∏–±–∫–∞: –í–≤–µ–¥–∏—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ —á–∏—Å–ª–∞")

window = Tk()
window.title("–ö–∞–ª—å–∫—É–ª—è—Ç–æ—Ä")
window.geometry('400x300')

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–∏–ª–µ–π
entry_style = {'width': 10, 'font': ('Arial', 12)}

# –í–≤–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–π x –∏ y
lblx = Label(window, text="x", font=('Arial', 14))
lblx.grid(column=0, row=0, pady=10)
txt1 = Entry(window, **entry_style)
txt1.grid(column=1, row=0, pady=10)
lbly = Label(window, text="y", font=('Arial', 14))
lbly.grid(column=2, row=0, pady=10)
txt2 = Entry(window, **entry_style)
txt2.grid(column=3, row=0, pady=10)

# –†–µ–∑—É–ª—å—Ç–∞—Ç
lbl = Label(window, text="–†–µ–∑—É–ª—å—Ç–∞—Ç - _", font=('Arial', 14))
lbl.grid(column=0, row=1, columnspan=4, pady=10)

# –ö–Ω–æ–ø–∫–∏ —Å –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞–º–∏
operators = ['+', '-', '*', '/', 'exp', 'sin', 'cos', '^']
row_number = 2
for operator in operators:
    if operator == 'sin' or operator == 'cos':
        btn_text = f"{operator}(x+y)"
    else:
        btn_text = f"x {operator} y"
    btn = Button(window, text=btn_text, command=lambda op=operator: calculate(op), font=('Arial', 12))
    btn.grid(column=operators.index(operator) % 4, row=row_number, padx=5, pady=5)
    if operators.index(operator) % 4 == 3:
        row_number += 1

# –ó–∞–ø—É—Å–∫ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è
window.mainloop()
```
